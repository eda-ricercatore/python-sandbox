## Generate Different Sine Waves and Plot Them

Define a sine wave, $y(t)$, as:

$y(t) = A \cdot sin(2 \cdot \pi \cdot f \cdot t + \varphi)$

such that:
+ $A$ is the amplitude of $y(t)$
+ $sin()$ is the sine function representing the sine wave $y(t)$
+ $\pi$ is the mathematical constant "pi"
+ $f$ is the frequency of $y(t)$
+ $t$ is the time instance of $y(t)$
+ $\varphi$ is the phase offset of $y(t)$


Generate sine waves for the following variations:
+ sine wave, $y_1(t) = A_{1} \cdot sin(2 \cdot \pi \cdot f \cdot t + \varphi)$, for a ***bigger amplitude, $A_{1} = 2A$***
+ sine wave, $y_2(t) = A \cdot sin(2 \cdot \pi \cdot f \cdot t + \varphi_{2})$, for a positive phase offset, $\varphi_{2} = \varphi + 200$
+ sine wave, $y_3(t) = A \cdot sin(2 \cdot \pi \cdot f \cdot t + \varphi_{3})$, for a positive phase offset, $\varphi_{3} = \varphi - 150$
+ sine wave, $y_4(t) = A \cdot sin(2 \cdot \pi \cdot f_{4} \cdot t + \varphi)$, for bigger frequency, $f_{4} = f \cdot \frac{4}{3}$

## Generate Different Square Waves and Plot Them

Define a parameterized square wave signal/process by the following properties:
+ frequency of the square wave
+ maximum amplitude of the square wave.
+ minimum amplitude of the square wave.


The square waves generated are:
+ $y(t)$, square wave with default values
+ $y_{1}(t)$, square wave with bigger frequency, $f_{1} = f \cdot \frac{3}{2}$
+ $y_{2}(t)$, square wave with bigger maximum amplitude, $A_{max_{2}} = A_{max} + 0.5$, and lesser/lower minimum amplitude, $A_{min_{2}} = A_{min} - 1.0$
+ $y_{3}(t)$, $y_{4}(t)$, and $y_{4}(t)$, square waves with the same maximum amplitude and minimum amplitude, but different from those of $y(t)$, and $y_{4}(t)$ and $y_{5}(t)$ have phase offsets.


Additional notes for the square waves $y_{3}(t)$, $y_{4}(t)$, and $y_{4}(t)$.
- $y_{4}(t)$ has a positive phase offset compared to $y_{3}(t)$, $\phi_{4} = \phi_{3} + 0.25$
    * A positive phase offset shifts the square wave leftwards.
- $y_{5}(t)$ has a negative phase offset compared to $y_{3}(t)$, 
$\phi_{5} = \phi_{3} - 0.40$
    * A negative phase offset shifts the square wave rightwards.
+ $A_{max_{3}} = A_{max_{4}} = A_{max_{5}} = A_{max}$
+ $A_{min_{3}} = A_{min_{4}} = A_{min_{5}} = A_{min} - 0.5$
